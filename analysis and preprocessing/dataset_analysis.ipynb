### Histogrami raspodele atributa

```python
h=pd.DataFrame.hist(dataset, figsize = (18,18), column = dataset.columns[:], color="red", bins=8, xlabelsize=12, ylabelsize=12)
```

```python
fig,(ax1, ax2) = plt.subplots(1, 2, figsize=(7,3.5))

ax1.hist(dataset["Month"])
ax1.set_title("Month")

ax2.hist(dataset["VisitorType"])
ax2.set_title("VisitorType")

plt.tight_layout()
plt.show()
```

### Broj klasa i njihove vrednosti atribura Revenue

```python
print('Broj klasa:', dataset['Revenue'].unique().shape[0])
for i in dataset['Revenue'].unique():
    print(i)
# dataset.Revenue.value_counts()
```

```python
print('Broj instanci klase True:', np.sum(dataset['Revenue'] == True))
print('Broj instanci klase Falase:', np.sum(dataset['Revenue'] == False))
```

### Procentualni prikaz balansa klasa 

```python
mlabels = "False", "True"
mcolors=["cyan", "magenta"]
mwedgeprops={"edgecolor":"black",'linewidth': 1.5}
p=plt.pie(dataset['Revenue'].value_counts(), labels=mlabels, colors=mcolors, autopct='%1.2f%%', wedgeprops = mwedgeprops)
```

### Odnos atributa Revenue i Month

```python
plt.figure(figsize=(15,5))
plt.title("Revenue vs Month")

months_order = ['Jan','Feb','Mar','Apr','May','June','Jul','Aug','Sep','Oct','Nov','Dec']
ax = sns.countplot(data=dataset, x='Month', hue='Revenue', order=months_order)
ax.legend(labels=['Not purchased','Purchased'])
for n in ax.containers:
    ax.bar_label(n)
    
plt.show()
```

### Odnos atributa Revenue i VisitorType

```python
plt.figure(figsize=(10,5))
plt.title("Revenue vs Visitor Type")

ax = sns.countplot(data=dataset, x='VisitorType', hue='Revenue')
ax.legend(labels=['Not purchased','Purchased'])
for n in ax.containers:
    ax.bar_label(n)

plt.show()
```

### Odnos atributa Revenue i SpecialDay

```python
plt.figure(figsize=(15,5))
plt.title("Revenue vs Special Day")

ax = sns.countplot(data=dataset, x='SpecialDay', hue='Revenue')
ax.legend(labels=['Not purchased','Purchased'])
for n in ax.containers:
    ax.bar_label(n)

plt.show()
```

### Odnos atributa Revenue i atributa tipova stranice 

```python
fig = plt.figure(figsize=(15, 10))

ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)

sns.violinplot(data=dataset, x='Revenue', y='Administrative', ax=ax1)
sns.violinplot(data=dataset, x='Revenue', y='Informational', ax=ax2)
sns.violinplot(data=dataset, x='Revenue', y='ProductRelated', ax=ax3)

plt.tight_layout()
plt.show()
```

### Odnos atributa Revenue i atributa vezanih za provedeno vrmeme na stranicama

```python
fig = plt.figure(figsize=(15, 10))

ax1 = fig.add_subplot(2, 3, 1)
ax2 = fig.add_subplot(2, 3, 2)
ax3 = fig.add_subplot(2, 3, 3)

sns.boxplot(data=dataset, x='Revenue', y='Administrative_Duration', ax=ax1)
sns.boxplot(data=dataset, x='Revenue', y='Informational_Duration', ax=ax2)
sns.boxplot(data=dataset, x='Revenue', y='ProductRelated_Duration', ax=ax3)

plt.tight_layout()
plt.show()
```
